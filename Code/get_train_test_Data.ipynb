```python
import pandas as pd
import json
from sklearn.model_selection import train_test_split
df = pd.read_excel("data.xlsx")
df["gap_eV"] = df["gap_eV"].apply(lambda x: f"{x:.3f}")
df["homo_eV"] = df["homo_eV"].apply(lambda x: f"{x:.3f}")
df["lumo_eV"] = df["lumo_eV"].apply(lambda x: f"{x:.3f}")
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
train_df, test_df = train_test_split(df, test_size=0.3, random_state=42)
def convert_to_jsonl(df, filename):
    with open(filename, "w", encoding="utf-8") as f:
        for _, row in df.iterrows():
            sample = {
                "input": f"分子的smiles表示: {row['smiles']}",
                "output": f"最高占据分子轨道能级:{row['homo_eV']};最低未占分子轨道能级:{row['lumo_eV']};带隙:{row['gap_eV']}"
            }
            f.write(json.dumps(sample, ensure_ascii=False) + "\n")
convert_to_jsonl(train_df, "./train_smi_homo_lumo_gap.jsonl")
convert_to_jsonl(test_df, "./test_smi_homo_lumo_gap.jsonl")
```
